## Matplotlib Install

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.arange(10)
```

## Interactive Graphics

```python
%matplotlib notebook
```

```python
plt.plot(data)
```

```python
plt.title("interactive test")
```

```python
plt.xlabel("sample name")
```

## Static Graphics

```python
%matplotlib inline
```

```python
x=[1,2,3,4]
y=[1,4,9,16]

plt.plot(x,y)
```

```python
d=np.arange(0.,5.,0.2)
```

```python
plt.plot(d, d,'r--',d,d**2,'bs',d,d**3,'g^')
```
